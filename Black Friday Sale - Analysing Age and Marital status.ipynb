```python
import pandas as pd
data = pd.read_csv('BlackFriday.csv')
```

```python
del data['Product_Category_2']
del data['Product_Category_3']
```

```python
data.info()
```

```python
data.groupby('Age').size().plot(kind = 'bar', figsize=(4,2), title = 'Purchase Distribution by Age' )
```

```python
data.groupby('Age').sum()['Purchase'].plot(kind = 'pie',autopct = '%0.1f')
```

```python
data.groupby('Age').sum()['Purchase'].plot(kind = 'bar', figsize = (4,2), title = 'Total Amount spend by Age Group')
```

```python
data.groupby('Age').mean()['Purchase'].plot(kind = 'bar', figsize = (4,2), title = 'Mean Amount spend by Age Group')
```

```python
data.head()
```

```python
lst = []
for i in data['Age'].unique():
    lst.append([i, data[data['Age']==i]['Product_ID'].nunique()])
df = pd.DataFrame(lst, columns = ['Age','Product'])
df.plot.bar(x='Age', title = 'Numberof products purchased by Age Group')
```

```python
data.groupby('Marital_Status').sum()['Purchase'].plot(kind = 'pie',autopct = '%0.1f')
```

```python
data.groupby('Gender').sum()['Purchase'].plot(kind = 'pie',autopct = '%0.1f')
```
