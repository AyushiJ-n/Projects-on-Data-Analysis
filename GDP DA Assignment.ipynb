## Task 2 : Create a new columns named as GDP which shows the GDP growth of each country for that year

```python
lst = []
old = 1
for i in data['Value']:
    new =  i
    growth = round((new-old)*100/old, 2)
    lst.append(growth)
    old = new
data['GDP Growth'] = lst
data.head()
```

## Task 3 : Create a graph of GDP Growth of India from 1960 - 2016

```python
data = data[data['Country Name']=='India']
data.plot(kind='line', x = 'Year', y = 'GDP Growth', grid = True, xlabel = 'YEAR', ylabel = 'GDP growth percentage' )
```
